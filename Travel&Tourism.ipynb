```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel files
df_arrivals = pd.read_csv("International_Tourist_Arrivals_2019_2023.csv")
df_fta = pd.read_csv("FTAs_in_India_2022_2023.csv")
```

```python
# Sum total international arrivals for each year
total_international_arrivals = {
    "2019": df_arrivals["2019_Arrivals (Million)"].sum(),
    "2022": df_arrivals["2022_Arrivals (Million)"].sum(),
    "2023": df_arrivals["2023_Arrivals (Million)"].sum()
}
```

```python

# Sum FTAs (Foreign Tourist Arrivals) in India
total_fta_india = {
    "2022": df_fta["FTAs_in_India_2022"].sum(),
    "2023": df_fta["FTAs_in_India_2023"].sum()
}
```

```python
# Chart 1: Global International Arrivals
# ---------------------------
plt.figure(figsize=(8, 5))
plt.bar(total_international_arrivals.keys(), total_international_arrivals.values(), color='skyblue')
plt.title("Global International Tourist Arrivals (in Million)")
plt.xlabel("Year")
plt.ylabel("Arrivals (Million)")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Chart 2: FTAs in India
# ---------------------------
plt.figure(figsize=(8, 5))
plt.bar(total_fta_india.keys(), total_fta_india.values(), color='lightgreen')
plt.title("Foreign Tourist Arrivals in India")
plt.xlabel("Year")
plt.ylabel("Number of Foreign Tourist Arrivals (In Million)")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
